```python
import numpy as np
import matplotlib.pyplot as plt

n_blockIdx_x = range(257)
n_threadIdx_x = range(64)
n_k = range(32)

memory_accesses_Apad = [b * 64 + t for b in n_blockIdx_x for t in n_threadIdx_x]
memory_accesses_W = [k for k in n_k]
memory_accesses_B = [b * 64 + t for b in n_blockIdx_x for t in n_threadIdx_x]

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))

# Plot for the Apad tensor
ax1.scatter(range(len(memory_accesses_Apad)), memory_accesses_Apad, color='red')
ax1.set_xlabel('Access Number')
ax1.set_ylabel('Memory Address')
ax1.set_title('Space-Time Plot for Apad Tensor (Apad_1)')

# Plot for the kernel tensor (W)
ax2.scatter(range(len(memory_accesses_W)), memory_accesses_W, color='green')
ax2.set_xlabel('Access Number')
ax2.set_ylabel('Memory Address')
ax2.set_title('Space-Time Plot for Kernel Tensor (W_1)')

# Plot for the output tensor (B)
ax3.scatter(range(len(memory_accesses_B)), memory_accesses_B, color='blue')
ax3.set_xlabel('Access Number')
ax3.set_ylabel('Memory Address')
ax3.set_title('Space-Time Plot for Output Tensor (B_1)')

plt.tight_layout()
plt.show()
```
